```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.utils import resample

# Cargar los datos con rutas absolutas
df0 = pd.read_csv(r'C:\DataScience\Sprint 11\geo_data_0.csv')
df1 = pd.read_csv(r'C:\DataScience\Sprint 11\geo_data_1.csv')
df2 = pd.read_csv(r'C:\DataScience\Sprint 11\geo_data_2.csv')

data = [df0, df1, df2]

# Explorar los datos
display("Región 0")
display(df0.info())
display(df0.describe())
display("-" * 40)

display("Región 1")
display(df1.info())
display(df1.describe())
display("-" * 40)

display("Región 2")
display(df2.info())
display(df2.describe())
display("-" * 40)

# Visualizar la distribución de las reservas de petróleo
df0['product'].hist(bins=30, alpha=0.5, label='Región 0')
df1['product'].hist(bins=30, alpha=0.5, label='Región 1')
df2['product'].hist(bins=30, alpha=0.5, label='Región 2')

plt.legend()
plt.xlabel('Volumen de reservas (miles de barriles)')
plt.ylabel('Frecuencia')
plt.title('Distribución de reservas de petróleo por región')
plt.show()
```

<b>Este código importa las librerías necesarias, carga los archivos CSV con datos geológicos de distintas regiones, y explora la estructura y estadísticas descriptivas de cada conjunto de datos. Luego, visualizar mediante histogramas la distribución del volumen de reservas (“producto”) en cada región</b> <a class="tocSkip"></a>
</div>

<div class="alert alert-block alert-success"> <b>Celda [1]</b> <a class="tocSkip"></a> Las librerías importadas son adecuadas para análisis de datos, modelado y visualización. Además, la carga de datos está correctamente implementada utilizando `pd.read_csv()`, permitiendo su posterior manipulación. </div> 

<div class="alert alert-block alert-success"> <b>Celda [1]</b> <a class="tocSkip"></a> La exploración inicial con `.info()` y `.describe()` es adecuada y proporciona información relevante sobre la estructura de los datos. También es positivo que se analicen las estadísticas de cada región de forma separada. </div> <div class="alert alert-block alert-warning"> <b>Celda [1]</b> <a class="tocSkip"></a> Podrías complementar la exploración de datos con análisis adicionales, como la detección de valores atípicos (`df.boxplot()`) y la correlación entre variables (`df.corr()`). Esto puede ayudar a identificar posibles relaciones entre las características y la variable objetivo. </div>

```python
# Función para entrenar y evaluar el modelo de regresión lineal
def train_and_evaluate(df, region_name):
    X = df[['f0', 'f1', 'f2']]
    y = df['product']
    
    # Dividir en entrenamiento (75%) y validación (25%)
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=42)
    
    # Crear el DataFrame de validación completo (con características y 'product')
    valid_data = X_valid.copy()
    valid_data['product'] = y_valid
    
    # Entrenar el modelo
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Hacer predicciones
    y_pred = model.predict(X_valid)
    
    # Calcular el RMSE
    rmse = mean_squared_error(y_valid, y_pred, squared=False)
    
    # Imprimir los resultados
    print(f'Resultados para {region_name}:')
    print(f'Volumen medio de reservas predicho: {y_pred.mean():.2f}')
    print(f'RMSE: {rmse:.2f}\n')
    
    # Devolver tanto las predicciones como el DataFrame de validación completo
    return model, y_pred, valid_data


# Entrenar y evaluar para cada región
model_0, pred_0, valid_0 = train_and_evaluate(df0, 'Región 0')
model_1, pred_1, valid_1 = train_and_evaluate(df1, 'Región 1')
model_2, pred_2, valid_2 = train_and_evaluate(df2, 'Región 2')
```

<div class="alert alert-block alert-info">
<b>En el código anterior define la función train_and_evaluate que extrae características y la variable objetivo, divide los datos en conjuntos de entrenamiento y validación, entrena un modelo de regresión lineal y realiza predicciones. Además, calcula el RMSE, imprime resultados y devuelve el modelo, las predicciones y el DataFrame de validación. Finalmente, se ejecuta para tres regiones.</b> <a class="tocSkip"></a>
</div>

<div class="alert alert-block alert-success"> <b>Celda [2]</b> <a class="tocSkip"></a> La implementación del modelo de regresión lineal es correcta. Se extraen las características y la variable objetivo, se dividen los datos en entrenamiento y validación, y se calculan predicciones y métricas de error. </div> <div class="alert alert-block alert-warning"> <b>Celda [2]</b> <a class="tocSkip"></a> Observé que se utiliza `squared=False` en `mean_squared_error()`, lo cual generará una advertencia (`FutureWarning`). Para evitar problemas con versiones futuras de `sklearn`, es recomendable cambiar a `root_mean_squared_error()` en versiones posteriores. </div>

```python
# Preparación para el cálculo de ganancias
WELL_COUNT = 200          # Número de pozos
REVENUE_PER_UNIT = 4500   # Ingresos por unidad (en dólares)
BUDGET = 100_000_000      # Presupuesto total
BREAKEVEN_PRODUCTION = BUDGET / WELL_COUNT / REVENUE_PER_UNIT  # Producción mínima necesaria

mean_reserves = [df['product'].mean() for df in data]

print("Punto de equilibrio (en miles de barriles por pozo):", BREAKEVEN_PRODUCTION)
print("Volumen medio de reservas por región:")
for i, mean_res in enumerate(mean_reserves):
    print(f'Región {i}: {mean_res:.2f}')

print("\nConclusión: Se debe considerar la región cuya media de reservas esté por encima del punto de equilibrio para maximizar las ganancias.")
```

<div class="alert alert-block alert-info">
<b>Este código define claves constantes (número de pozos, ingresos por unidad y presupuesto) y calcula la producción mínima necesaria para alcanzar el punto de equilibrio. Además, determina el volumen medio de reservas por región, imprimiendo ambos valores y concluyendo que se deben priorizar las regiones con reservas superiores al punto de equilibrio</b> <a class="tocSkip"></a>
</div>

<div class="alert alert-block alert-success"> <b>Celda [3]</b> <a class="tocSkip"></a> El cálculo del punto de equilibrio es correcto y proporciona una base para la toma de decisiones basada en la producción mínima necesaria por pozo. </div> <div class="alert alert-block alert-warning"> <b>Celda [3]</b> <a class="tocSkip"></a> Sin embargo, la conclusión basada en la comparación de medias puede ser limitada. Sería útil analizar también la distribución de reservas en cada región y considerar la desviación estándar, ya que una alta variabilidad en las reservas puede afectar la rentabilidad real. </div>

```python
# Seleccionamos los 200 pozos con las mayores reservas en cada región y calculamos la ganancia esperada.
def calculate_profit(predictions, valid_data, region_name):
    selected_indices = predictions.argsort()[-WELL_COUNT:][::-1]
    selected_reserves = valid_data.iloc[selected_indices]['product'].sum()
    revenue = selected_reserves * REVENUE_PER_UNIT
    profit = revenue - BUDGET
    print(f'Ganancia estimada para {region_name}: ${profit:,.2f}')
    return profit

# Calcular ganancias por región
profit_0 = calculate_profit(pred_0, valid_0, 'Región 0')
profit_1 = calculate_profit(pred_1, valid_1, 'Región 1')
profit_2 = calculate_profit(pred_2, valid_2, 'Región 2')

# Determinar la mejor región
profits = [profit_0, profit_1, profit_2]
best_region = np.argmax(profits)
print(f'La mejor región para perforar es la Región {best_region} con una ganancia estimada de ${profits[best_region]:,.2f}')
```

<div class="alert alert-block alert-info">
<b>Esta función selecciona los 200 pozos con las predicciones más altas, suma sus reservas reales (columna “producto”) y calcula los ingresos multiplicando por el valor unitario, restando el presupuesto para obtener la ganancia. Se ejecuta para cada región y se comparan las ganancias para identificar la mejor. Claramente la Región 0 es la que promete ser la más apropiada para escoger.</b> <a class="tocSkip"></a>
</div>

<div class="alert alert-block alert-success"> <b>Celda [4]</b> <a class="tocSkip"></a> La lógica para seleccionar los 200 pozos con mayor reserva y calcular la ganancia esperada está bien estructurada. Se utiliza `argsort()` para obtener los índices de los pozos con mayor predicción, lo cual es una estrategia válida. </div> <div class="alert alert-block alert-warning"> <b>Celda [4]</b> <a class="tocSkip"></a> Podrías añadir un chequeo para asegurarte de que los pozos seleccionados efectivamente superen el umbral del punto de equilibrio. Esto podría evitar seleccionar regiones con alta variabilidad y un riesgo mayor de pérdida. </div>

```python
# Función para entrenar y evaluar el modelo
def train_and_evaluate(df, region_name):
    X = df[['f0', 'f1', 'f2']]
    y = df['product']
    
    # Dividir en entrenamiento (75%) y validación (25%)
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=42)
    
    # Crear el DataFrame de validación completo (con características y 'product')
    valid_data = X_valid.copy()
    valid_data['product'] = y_valid
    
    # Entrenar el modelo
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Hacer predicciones
    y_pred = model.predict(X_valid)
    
    # Calcular el RMSE
    rmse = mean_squared_error(y_valid, y_pred, squared=False)
    
    # Imprimir los resultados
    print(f'Resultados para {region_name}:')
    print(f'Volumen medio de reservas predicho: {y_pred.mean():.2f}')
    print(f'RMSE: {rmse:.2f}\n')
    
    # Devolver tanto las predicciones como el DataFrame de validación completo
    return model, y_pred, valid_data
```

<div class="alert alert-block alert-info">
<b>Esta Función entrena un modelo de regresión lineal para predecir el volumen de reservas de petróleo en diferentes regiones. Divide los datos en entrenamiento y validación, entrena el modelo, genera predicciones y calcula el error cuadrático medio (RMSE). Finalmente, imprime los resultados y retorna el modelo, las predicciones y los datos de validación.</b> <a class="tocSkip"></a>
</div>

```python
# Función de bootstrapping
def bootstrap_profit(predictions, valid_data, region_name, n_iterations=1000, confidence_level=0.95):
    profits = []
    
    # Combinar valid_data y las predicciones en un DataFrame
    df_combined = valid_data.copy()
    df_combined['prediction'] = predictions
    
    for _ in range(n_iterations):
        # Muestreo con reemplazo del DataFrame combinado
        sampled_data = resample(df_combined, replace=True, n_samples=len(df_combined))
        
        # Ordenar por la columna 'prediction' (de mayor a menor)
        sampled_data = sampled_data.sort_values(by='prediction', ascending=False)
        
        # Seleccionar los WELL_COUNT (200) mejores pozos y sumar la columna 'product'
        selected_reserves = sampled_data.head(WELL_COUNT)['product'].sum()
        
        # Calcular la ganancia para esta muestra
        revenue = selected_reserves * REVENUE_PER_UNIT
        profit = revenue - BUDGET
        profits.append(profit)
    
    # Calcular el beneficio medio y el intervalo de confianza
    mean_profit = np.mean(profits)
    lower_bound = np.percentile(profits, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(profits, (1 + confidence_level) / 2 * 100)
    
    # Calcular el riesgo de pérdida (proporción de iteraciones con ganancia negativa)
    risk_of_loss = np.mean(np.array(profits) < 0)
    
    return mean_profit, lower_bound, upper_bound, risk_of_loss
```

<div class="alert alert-block alert-info">
<b>Este código implementa una función de bootstrapping para estimar la ganancia de un proyecto de perforación. Realice muestras con reemplazo, calcule la ganancia para cada muestra y repita el proceso varias veces. Al final, calcule el beneficio medio , el intervalo de confianza y el riesgo de pérdida</b> <a class="tocSkip"></a>
</div>

<div class="alert alert-block alert-success"> <b>Celda [6]</b> <a class="tocSkip"></a> La implementación del bootstrapping es adecuada y permite una estimación más robusta de las ganancias esperadas y su intervalo de confianza. Es positivo que se calcule el `riesgo de pérdida`, lo que añade una capa adicional de análisis. </div> <div class="alert alert-block alert-warning"> <b>Celda [6]</b> <a class="tocSkip"></a> Sin embargo, el número de iteraciones (`n_iterations=1000`) podría ser configurado como un parámetro opcional para permitir ajustes en función de los recursos computacionales disponibles. También sería útil visualizar la distribución de los beneficios con un histograma para observar la variabilidad de las simulaciones. </div>

```python
# Entrenar y evaluar para cada región
model_0, pred_0, valid_data_0 = train_and_evaluate(df0, 'Región 0')
model_1, pred_1, valid_data_1 = train_and_evaluate(df1, 'Región 1')
model_2, pred_2, valid_data_2 = train_and_evaluate(df2, 'Región 2')

# Realizar el análisis de bootstrapping para cada región
result_0 = bootstrap_profit(pred_0, valid_data_0, 'Región 0')
result_1 = bootstrap_profit(pred_1, valid_data_1, 'Región 1')
result_2 = bootstrap_profit(pred_2, valid_data_2, 'Región 2')

# Almacenar y comparar los resultados
results = {
    'Región 0': result_0,
    'Región 1': result_1,
    'Región 2': result_2
}

for region, result in results.items():
    mean_profit, lower_bound, upper_bound, risk_of_loss = result
    print(f"Resultados para {region}:")
    print(f"Beneficio medio: ${mean_profit:,.2f}")
    print(f"Intervalo de confianza (95%): ${lower_bound:,.2f} a ${upper_bound:,.2f}")
    print(f"Riesgo de pérdida: {risk_of_loss * 100:.2f}%\n")

best_region = min(results, key=lambda region: (results[region][0], results[region][3]))
print(f"La mejor región para perforar es {best_region}.")
```

Posteriormente, compare el beneficio medio , el intervalo de confianza y el riesgo de pérdida para cada región.
